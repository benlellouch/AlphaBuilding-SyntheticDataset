# Time Lagged Correlations

```python
def crosscorr(datax, datay, lag=0):
    """ Lag-N cross correlation. 
    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length

    Returns
    ----------
    crosscorr : float
    """
    return datax.corr(datay.shift(lag))
```

```python
def plot_crosscorr(rs, ):
    f,ax=plt.subplots(figsize=(14,3))
    ax.plot(rs)
    ax.axvline(np.ceil(len(rs)/2),color='k',linestyle='--',label='Center')
    ax.axvline(np.argmax(rs),color='r',linestyle='--',label='Peak synchrony')
    plt.legend()
```

### Comparing Occupancy to MELs

```python
rs = [crosscorr(ts_1a_lo['Total Occupant Count'], ts_1a_lo['MELs (kWh)'], lag=i) for i in range(-6,6)]
plot_crosscorr(rs)
```

```python

rs = [crosscorr(ts_1a_hi['Total Occupant Count'], ts_1a_hi['MELs (kWh)'], lag=i) for i in range(-6,6)]
plot_crosscorr(rs)
```
